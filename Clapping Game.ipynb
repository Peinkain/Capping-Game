**Rules of the Clapping Game** <br>
As every drinking game, this one has also very simple rules. Create a circle with a group of friends. The first player chooses any integer. (Choose direction either clock-wise or counter clock-wise). The next person must say the following number if it doesn't contain 3,6 or 9. If it does contain one of these numbers, the player must clap his hands instead. Thus, the number is basically skipped and it's next person's turn. If these numbers appear more than once in the following integer (e.g 36), then the player must clap as many times as the numbers appear in the integer (in this example case it would be twice).

## Libraries ##
For this game, we don't do any further analysis. Thus, simple NumPy will be sufficient for all numerical calcualtions.

```python
import numpy as np
```

```python
starting_number = 7                # the number that was chosen by the first player
number_of_players = 5              # number of players playing the game
my_turn = 3                        # your place relative to the first player

numbers = np.arange(1,100)         # number of turns
N = len(numbers)                   

en_1 = '3'                         # Forbidden numer 3
en_2 = '6'                         # Forbidden numer 6
en_3 = '9'                         # Forbidden numer 9

my_numbers = starting_number + (my_turn - 1) + number_of_players * (numbers - 1)    # numbers you will have to say or clap

n = len(my_numbers)
```

Result of the cell below are printed numbers you will have to say or clap each round with number of claps included.

```python
for i in range(n):
    if en_1 in str(my_numbers[i]) and en_2 in str(my_numbers[i]) and en_3 in str(my_numbers[i]):
        print('clap clap clap' + f'[{my_numbers[i]}]')
    elif en_1 in str(my_numbers[i]) and en_2 in str(my_numbers[i]):
        print('clap clap' + f'[{my_numbers[i]}]')
    elif en_2 in str(my_numbers[i]) and en_3 in str(my_numbers[i]):
        print('clap clap' + f'[{my_numbers[i]}]')
    elif en_3 in str(my_numbers[i]) and en_1 in str(my_numbers[i]):
        print('clap clap' + f'[{my_numbers[i]}]')
    elif en_1 in str(my_numbers[i]):
        string = str(my_numbers[i])
        print(string.count(en_1) * 'clap ' + f'[{my_numbers[i]}]')
    elif en_2 in str(my_numbers[i]):
        string = str(my_numbers[i])
        print(string.count(en_2) * 'clap ' + f'[{my_numbers[i]}]')
    elif en_3 in str(my_numbers[i]):
        string = str(my_numbers[i])
        print(string.count(en_3) * 'clap ' + f'[{my_numbers[i]}]')
    else:
        print(my_numbers[i])
```
